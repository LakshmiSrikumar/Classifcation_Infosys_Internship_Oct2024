```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Step 1: Data Preparation (Assuming you have organized train/validation datasets)
# Use ImageDataGenerator to preprocess images
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)

validation_datagen = ImageDataGenerator(rescale=1.0 / 255)

train_generator = train_datagen.flow_from_directory(
    'data/train',  # Path to train dataset
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

validation_generator = validation_datagen.flow_from_directory(
    'data/validation',  # Path to validation dataset
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

# Step 2: Build the CNN Model
model = Sequential()

# Input and Convolutional Layer
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(150, 150, 3)))

# Pooling Layer
model.add(MaxPooling2D(pool_size=(2, 2)))

# Additional Convolutional Layer
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Fully Connected Layer
model.add(Flatten())
model.add(Dense(units=128, activation='relu'))

# Output Layer
model.add(Dense(units=2, activation='softmax'))

# Step 3: Compile the Model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Step 4: Train the Model
model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=10,  # Number of epochs
    validation_data=validation_generator,
    validation_steps=len(validation_generator)
)

# Step 5: Evaluate or Predict
# Test with a single image
from tensorflow.keras.preprocessing import image
import numpy as np

# Replace with the path to your test image
test_image_path = r'C:\Users\yamun\OneDrive\Desktop\Traine'
img = image.load_img(test_image_path, target_size=(150, 150))
img_array = image.img_to_array(img) / 255.0  # Normalize
img_array = np.expand_dims(img_array, axis=0)

# Predict
prediction = model.predict(img_array)
classes = train_generator.class_indices
class_labels = {v: k for k, v in classes.items()}
predicted_class = class_labels[np.argmax(prediction)]

print(f"Predicted Class: {predicted_class}")
```
